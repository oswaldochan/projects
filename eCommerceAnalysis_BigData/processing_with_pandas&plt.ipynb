## eCommerce Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

## Heat ðŸ”¥ map ðŸ—º

```python
heatmap = pd.read_csv('HeatMap.csv', header=None, names=['day', 'hour', 'purchases_count'])
heatmap = heatmap.sort_values(by=['day', 'hour']).reset_index(drop=True);
heatmap = heatmap.fillna(0)
```

```python
hm = heatmap.pivot_table(values='purchases_count', index='hour', columns='day')
```

```python
_, __ = plt.subplots(figsize=(15, 12))
# to show the format as num + h
sns.heatmap(hm, yticklabels=[str(i)+' h' for i in range(0,24)], cmap='OrRd')
plt.title('NUMBER OF PURCHASES BY HOUR AND DAY', fontsize=30, pad=2)
plt.savefig('heatmap.png')
```

## Hist 1-5

```python
# smartphone
hist1 = pd.read_csv('Hist1.csv', header=None, usecols=[6], names=['smartphone'], squeeze=True)

# headphone
hist2 = pd.read_csv('Hist2.csv', header=None, usecols=[6], names=['headphone'], squeeze=True)

# tv
hist3 = pd.read_csv('Hist3.csv', header=None, usecols=[6], names=['tv'], squeeze=True)

# clocks
hist4 = pd.read_csv('Hist4.csv', header=None, usecols=[6], names=['clocks'], squeeze=True)

# washer
hist5 = pd.read_csv('Hist5.csv', header=None, usecols=[6], names=['washer'], squeeze=True)
```

```python
grid = plt.GridSpec(4, 6, wspace=0.4, hspace=0.4)
fig = plt.figure(figsize=(12, 8))

principal = fig.add_subplot(grid[:, 0:2])
sns.distplot(hist1, color='red', ax=principal, axlabel='#1 - Smartphone')

# ------------

arriba_izquierda = fig.add_subplot(grid[:2, 2:4])
sns.distplot(hist2, ax=arriba_izquierda, axlabel='#2 - Headphone')
plt.title('PRICE DISTRIBUTION IN DOLLARS OF THE TOP 5 CATEGORIES', loc='center', pad=1.3, fontsize=20);


arriba_derecha = fig.add_subplot(grid[:2, 4:])
sns.distplot(hist3, ax=arriba_derecha, axlabel='#3 - TV')

# -----------

abajo_izquierda = fig.add_subplot(grid[2:, 2:4])
sns.distplot(hist4, ax=abajo_izquierda, axlabel='#4 - Clocks')

abajo_derecha = fig.add_subplot(grid[2:, 4:])
sns.distplot(hist2, ax=abajo_derecha, axlabel='#5 - Washer')

# Hide y-axis
arriba_derecha.axes.get_yaxis().set_visible(False)
abajo_derecha.axes.get_yaxis().set_visible(False)
arriba_izquierda.axes.get_yaxis().set_visible(False)
abajo_izquierda.axes.get_yaxis().set_visible(False)
```

## Horizontal bar plot

```python
# purchase
hbp1 = pd.read_csv('HorizontalBarPlot1.csv', header=None, index_col=0, names=['purchase'], squeeze=True).sort_index()

# cart
hbp2 = pd.read_csv('HorizontalBarPlot2.csv', header=None, index_col=0, names=['cart'], squeeze=True).sort_index()

# views
hbp3 = pd.read_csv('HorizontalBarPlot3.csv', header=None, index_col=0, names=['views'], squeeze=True).sort_index()
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(17, 15))


# Plot the total cart additions
sns.set_color_codes("deep")
sns.barplot(hbp2.values, hbp2.index,
            label="Total 2 ", color="b", orient='h')


# Plot the total purchases
sns.set_color_codes("bright")
sns.barplot(hbp1.values, hbp1.index,
            label="Total 1", color="b", orient='h')


# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Added to cart vs Sold")
sns.despine(left=True, bottom=True)
```

## Stacked

## Word cloud

```python
import pandas as pd

wc = pd.read_csv('WordNubeVENTAS.csv', header=None, names=['cat', 'value'])

# Para extraer la Ãºltima palabra de categorÃ­a
wc.cat = wc.cat.apply(lambda x: x.split('.')[-1])

categorias = list(wc.cat)
valores = list(wc.value)
```

```python
# Libraries
from wordcloud import WordCloud
 
# Create a list of word
text= ' '.join(categorias)
 
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0,background_color='white', colormap='plasma').generate(text)
 
# Display the generated image:
plt.figure(figsize=[6,6])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('wordcloud_square', dpi=150, transparent=True)
plt.show()

# Reference https://python-graph-gallery.com/261-custom-python-wordcloud/
```

```python
import numpy as np
from PIL import Image    # to import the image
# Load the image (http://python-graph-gallery.com/wp-content/uploads/wave.jpg)
wave_mask = np.array(Image.open( "compu.jpg"))
 
# Make the figure
wordcloud = WordCloud(mask=wave_mask, background_color='white', colormap='gnuplot').generate(text)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('wordcloud_pc', dpi=150, transparent=True)
plt.show()
```
